#Tokenization

```python
import nltk
nltk.download("punkt_tab")
```

```python
sentence = """The Supreme Court on Monday gave a verdict that changed Priya's life."""
```

```python
tokens = nltk.word_tokenize(sentence)
```

```python
sentence
```

```python
tokens
```

```python
nltk.sent_tokenize(sentence)
```

#Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for t in tokens:
  print(t, " : ", ps.stem(t))
```

#Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemma = WordNetLemmatizer()
```

```python
for t in tokens:
  print(t, " : ", lemma.lemmatize(t))
```

#Identify & Remove Stopwords



```python
from nltk.corpus import stopwords
nltk.download("stopwords")

sw = set(stopwords.words("english"))

filtered_tokens = [t for t in tokens if t.lower() not in sw]

print(filtered_tokens)
```
